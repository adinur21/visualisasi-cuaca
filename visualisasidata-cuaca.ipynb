```python
# import libraries
import pandas as pd 
import matplotlib.pyplot as plt  

%matplotlib inline
```

```python
weather = pd.read_csv("weatherlondon.csv")
weather.info()
```

```python
weather.head()
```

```python
# simple plot : (by default create line chart) Line Chart
weather.plot(y='Tmax', x='Month')
plt.show()
```

```python
# Multiple Line Chart
weather.plot(y=['Tmax','Tmin'], x='Month')
plt.show()
```

```python
weather['Tmed'] = (weather['Tmax'] + weather['Tmin'])/2
weather
```

```python
weather.plot(y=['Tmax','Tmin', 'Tmed'], x='Month')
plt.show()
```

```python
# Bar Chart
weather.plot(kind='bar', y='Rain', x='Month')
plt.show()
```

```python
#same result would be obtained if we use specific bar chart method :
weather.plot.bar(y='Rain', x='Month')
plt.show()
```

```python
weather.plot(kind='barh', y='Rain', x='Month')
plt.show()
```

```python
weather.plot(kind='bar', y=['Tmax','Tmin'], x='Month')
plt.show()
```

```python
# scatter diagram
weather.plot(kind='scatter', x='Sun', y='Rain')
plt.show()
```

```python
# pie chart
weather.plot(kind='pie', y='Sun')
plt.show()
```

```python
weather['idxMonth'] =['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
weather
weather.set_index('idxMonth', inplace=True)
#weather.index=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
weather.info()
weather.head()
```

```python
weather.plot.pie(y = 'Sun', legend=False)
# or:
# weather.plot(kind='pie', y = 'Sun', legend=False)
plt.show()
```

```python
weather.plot.pie(y = 'Sun', legend=False, figsize=(7,7), autopct='%1.1f%%')
plt.show()
```

```python
# multiple chart using subplot=True
weather.plot(y=['Tmax', 'Tmin','Rain','Sun'], subplots=True, layout=(2,2), figsize=(10,5))
plt.show()
```

```python
# multiple bar chart
weather.plot(kind='bar', y=['Tmax', 'Tmin','Rain','Sun'], subplots=True, layout=(2,2), figsize=(10,5))
plt.show()
```

```python
#create new dataframe consists columns we need to plot into multiple pie chart
weather2 = weather[['Tmax', 'Tmin','Tmed', 'Rain','Sun']]
weather2
```

```python
# multiple pie chart of Tmax, Tmin, Tmed, Rain, Sun
weather2.plot(kind='pie', subplots=True, legend=False, layout=(3,2),figsize=(10,15))
plt.show()
```

```python
# saving the chart
weather.plot(kind='bar', y='Rain', legend=False)
plt.savefig('bar.jpeg')
plt.show()
```

```python
weather2.plot(kind='pie', subplots=True, legend=False, layout=(2,3),figsize=(15,10))
plt.savefig('D:/pychart/multichart_pie.png')
plt.show()
```

```python
# line 1 points
x1 = [10,20,30]
y1 = [20,40,10]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")

# line 2 points
x2 = [10,20,30]
y2 = [40,10,30]
# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")

plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Multiple Lines')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

```python
plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-width-3')
plt.plot(x2,y2, color='red', linewidth = 5,  label = 'line2-width-5')
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Multiple Lines')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
